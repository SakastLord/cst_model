## CONT STOIKOV TALREJA BOOK SHAPE

```python
L = 30
lamda_emp = np.array([1.85,1.51,1.09,0.88,0.77])
lamdas = np.append(lamda_emp,cst_book.powerlawfit(lamda_emp,np.arange(6,L+1)))
thetas = np.append([0.71,0.81,0.68,0.56,0.47],np.repeat(0.47,L-5))
x = Book()
y = Book()
z = Book()
mu = 0.94
```

```python
result = x.cst_simulate(mu,lamdas,thetas,L,10**4)
```

```python
resul2 = y.cst_simulate(mu,lamdas,thetas,L,10**5)
```

```python
# very slow to run
result3 = z.cst_simulate(mu,lamdas,thetas,L,5*10**5)
```

```python
import pandas as pd
```

```python
col_events = ['$10^4$ events','$10^5$ events','$10^6$ events']
results = pd.DataFrame(np.column_stack([result,resul2,result3]),columns=col_events)
results['$10^4$ events'].plot(figsize=(8,4),color='green')
results['$10^5$ events'].plot(color='orange');
results['$10^6$ events'].plot(color='c');
plt.xlabel('distance (ticks) from opposite best quote');
plt.ylabel('average quantities');
plt.title('Average Book Shape in CST Model');
plt.legend(loc='best');
```
